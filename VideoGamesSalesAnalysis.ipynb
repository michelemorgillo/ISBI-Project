# Predictive modeling

Utilizzo il DataFrame df_sales per creare un modello di previsione.

```python
df_sales
```

Creo variabili dummy per le caratteristiche di tipo oggetto. Utilizzo il dataframe df_sales per creare un modello, escludendo la variabile "Name" e altre caratteristiche critiche.

```python
df_sales_dummies = pd.get_dummies(df_sales.iloc[:,2:11], drop_first=True)
df_sales_dummies.columns
```

Ottengo quindi 625 caratteristiche dummy.

Divido i dati e normalizzo le caratteristiche usando il modello StandardScaler di sklearn.

```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler

X = df_sales_dummies.drop('Global_Sales', axis=1).values
y = df_sales_dummies['Global_Sales'].values

#divido i dati in testing e training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 55)

#normalizzo
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Utilizzo il modello PCA per capire quanto contribuisce ciascuna componente principale alla variazione complessiva nei dati. È utile per comprendere quante componenti siano necessarie per catturare la maggior parte della variazione nei dati originali. Idealmente, si cerca di scegliere un numero ottimale di componenti principali che spieghino una grande parte della varianza, riducendo al contempo la dimensionalità dei dati.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=60) #provo diversi n_components per vedere il risultato
df_pca = pca.fit(X_train_scaled)
features = range(pca.n_components_)

#istogramma della varianza
plt.bar(features, pca.explained_variance_ratio_)
plt.ylabel('variance')
plt.xlabel('PCA feature')
plt.show()
```

La varianza rimane relativamente stabile dopo circa il 40° componente. Pertanto, scelgo 40 componenti.

Successivamente eseguo la riduzione della dimensionalità dei dati sia per il set di addestramento che per quello di test utilizzando il modello PCA precedentemente addestrato con 40 componenti principali. La riduzione della dimensionalità può aiutare a gestire meglio i dati e migliorare le prestazioni del modello, riducendo la complessità senza perdere molte informazioni importanti.

```python
pca = PCA(n_components=40)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

Con il seguente codice addestro un modello di regressione lineare (LinearRegression) sulle componenti principali ottenute dal PCA, esegue previsioni sui dati di test e calcola l'RMSE per valutare le prestazioni del modello. L'RMSE fornisce una stima della precisione del modello rispetto ai dati di test utilizzati. Un RMSE più basso indica una maggiore accuratezza delle previsioni.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train_pca, y_train)
y_pred = model.predict(X_test_pca)

rmse = mean_squared_error(y_test, y_pred, squared=False)
print('Root Mean Squared Error (RMSE) : '+str(round(rmse,4)))
```

 Eseguo la validazione incrociata (cross-validation) utilizzando la tecnica K-Fold con 10 fold sul modello di regressione lineare precedentemente addestrato. Questa fornisce una stima della performance media del modello su dati non visti. Aiuta a valutare la capacità del modello di generalizzare su nuovi dati, senza essere influenzato eccessivamente dalla particolare suddivisione dei dati di addestramento e test.

```python
kf = KFold(n_splits=10, random_state=55, shuffle=True) #divido i dati in 10 fold
cv_score = cross_val_score(model, X_train_pca, y_train, cv=kf)
print('Cross Validation Score : '+str(round(cv_score.mean(),4)))
```

```python
#scatter plot del modello di regressione
plt.scatter(y_test, y_pred)
plt.xlabel("Valori Veri")
plt.ylabel("Valori Predetti")
plt.title(f'Regression Model - RMSE: {rmse:.2f}')
plt.show()
```

Dai risultati ottenuti, emerge che l'RMSE (Root Mean Squared Error) per la previsione delle vendite globali rimane piuttosto elevato, intorno a 0,4 milioni. Tuttavia, si registra un'alta precisione. Nella nostra ricerca di un RMSE basso per questo modello di regressione, dobbiamo concentrarci sull'ottimizzazione del modello per ridurre ulteriormente l'errore.

Ecco alcune osservazioni chiave:

* Il dataset contiene dati storici che risalgono a circa 30 anni fa, in contrasto con una disponibilità limitata di dati recenti. Questa disparità può costituire una sfida per l'accuratezza del modello. L'abbondanza di dati storici potrebbe influenzare notevolmente il processo di apprendimento, mentre la carenza di dati recenti potrebbe compromettere la capacità del modello di catturare le tendenze più recenti.
* Le singole righe nel dataset mancano di informazioni cruciali che influenzano le vendite, come il genere del gioco, la valutazione grafica, la storia dei punteggi, il prezzo del bundle e altri fattori rilevanti.
* Nonostante il dataset includa valutazioni delle recensioni, i tentativi di ottimizzazione del modello hanno dimostrato che l'inclusione di tali caratteristiche porta ad un aumento dell'RMSE del modello.
